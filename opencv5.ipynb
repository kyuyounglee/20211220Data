```python
import cv2
import numpy as np
```

```python
#15%를 알파 블랜딩의 범위로지정
alpha_width_rate = 15
```

```python
img_face = cv2.imread('d:/man_face.jpg')
img_skull = cv2.imread('d:/skull.jpg')
```

```python
# 입력영상과 같은 크기의 영상 준비... 결과이미지를 저장할..
img_comp =  np.zeros_like(img_face)
```

```python
img_face.shape
```

```python
img_face.shape[:2]
```

```python
height,width = img_face.shape[:2]
middle = width // 2
alpha_width = width + alpha_width_rate // 100  # 알파블랜딩의 범위
start = middle - alpha_width//2  # 알파블랜딩의 시작지점
step = 100 / alpha_width        # 알파값 간격
```

```python
# 절반씩 복사.. 결과에 합성
img_comp[:,:middle,:] = img_face[:,:middle,:].copy()
img_comp[:,middle:,:] = img_skull[:,middle:,:].copy()
cv2.imshow('half', img_comp)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 알파값을 변경하면서 블렌딩 적용
for i in range(alpha_width):
    alpha = (100-step*i) / 100  # 알파값을 (0~1)
    beta = 1- alpha
    #알파 블랜딩 
    img_comp[:,start+i] = img_face[:,start+i]*alpha+img_skull[:,start+i]*beta
    print(i,alpha,beta)
    
cv2.imshow('half skull', img_comp)    
cv2.waitKey()
cv2.destroyAllWindows()
```
